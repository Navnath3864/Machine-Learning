```python
def getIrisData(index,df_iris):
    print("\n sepel length: ",df_iris.data[index][0])
    print("\n sepel width: ",df_iris.data[index][1])
    print("\n petal length: ",df_iris.data[index][2])
    print("\n petal width: ",df_iris.data[index][3])
    print("\n category: ",df_iris.target_names[df_iris.target[index]])
```

```python
def visualizeData(model,iris):
    model.fit(iris.data,iris.target)
    
    import matplotlib.pyplot as plt
    fig=plt.figure(figsize=(25,20))

    from sklearn import tree

    tree.plot_tree(model,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
        #import the tree class
    
```

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df_iris=datasets.load_iris()
```

```python
x=df_iris.data
y=df_iris.target
```

```python
#encode the data using StandardScaler
en_sc=StandardScaler()
x=en_sc.fit_transform(x)
```

```python
#split the data into train and test part
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#initialize our decision tree object
model=DecisionTreeClassifier(criterion="gini")
```

```python
#visualize the data with decision tree algorithms
visualizeData(model,df_iris)
getIrisData(1,df_iris)
getIrisData(55,df_iris)
getIrisData(120,df_iris)
```

```python
#train the model using x and y values
model.fit(x_train,y_train)
```

```python
#get the testing data predict it
labels=model.predict(x_test)
```

```python
#test the model
cm=metrics.confusion_matrix(y_test,labels)
print("accuracy : ",metrics.accuracy_score(y_test,labels)*100)
```
